```python
#Specify values for each column.
df1 = pd.DataFrame(
{"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])
display(df1)
```

```python
#Specify values for each row.
df2 = pd.DataFrame(
[[4, 7, 10],
[5, 8, 11],
[6, 9, 12]],
index=[1, 2, 3],
columns=['a', 'b', 'c'])
display(df2)
```

```python
#Create DataFrame with a MultiIndex
df3 = pd.DataFrame(
{"a" : ["a" ,5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = pd.MultiIndex.from_tuples(
[("d", 1), ("d", 2),
('e', 2)], names=['n', 'v']))
display(df3)
```

```python
#Most pandas methods return a DataFrame so that
#another pandas method can be applied to the result.
#This improves readability of code.
import pandas as pd
search_text = 'val == a'

# Assuming df3 is your DataFrame
df3_melted = pd.melt(df3).rename(columns={'variable': 'a', 'value': 'val'})

# Filter rows where the value in column 'val' is equal to 1 in column 'a'
df3_filtered = df3_melted.query(search_text)

# Display the resulting DataFrame
display(df3_filtered)
```
