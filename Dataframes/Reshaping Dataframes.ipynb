```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
{"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])
display(df1)

df2 = pd.DataFrame(
{"a" : [11, 11, 11],
"b" : [11, 11, 11],
"c" : [11, 11, 11]},
index = [1, 2, 3])
display(df2)
```

```python
#Append rows of DataFrames
df3 = pd.DataFrame(pd.concat([df1,df2]))
display(df3)
```

```python
#Spread rows into columns.

# Sample DataFrame (replace this with your actual DataFrame)
data = {'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9]}
df3 = pd.DataFrame(data)

# Melt the DataFrame to have 'var' and 'val' columns
df3_melted = pd.melt(df3, var_name='var', value_name='val')

# Pivot the melted DataFrame
df3_pivoted = df3_melted.pivot(columns='var', values='val')

# Display the resulting pivoted DataFrame
display(df3_pivoted)
```
